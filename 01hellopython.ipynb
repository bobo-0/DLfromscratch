```python
class Man:
    def __init__(self, name):
        self.name = name
        print("initialized!")

    def hello(self):
        print("Hello "+ self.name + "!")
    
    def goodbye(self):
        print("Good-bye "+ self.name + "!")

m = Man("David")
m.hello()
m.goodbye()
```

```python
X = np.array([[51, 55], [14,19],[0,4]])
print(X)

print(X[0])

print(X[0][1])

for row in X:
	print(row)
```

```python
#X를 1차원 배열로 평단화
X = X.flatten()
print(X)

#인덱스가 0, 2, 4인 원소를 얻기
X[np.array([0,2,4])]

#비교 연산을 bool 결과로 얻기, 크면 True 작으면 False
print(X >15)

#15보다 큰 원소만 꺼내기
print(X[X>15])
```

```python
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1, label = "sin")
plt.plot(x,y2, linestyle="--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```
